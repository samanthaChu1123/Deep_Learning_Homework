```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(-7.5, 7.5, 0.3)
y1 = 4 * (x+6) * (x+1) * (x-3) * (x-7)
y2 = 3 * (x+1) * (x-5) * (x+3)
y3 = 100 * (x+1) * (x-6)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(['f(x)','g(x)','k(x)'])
```
